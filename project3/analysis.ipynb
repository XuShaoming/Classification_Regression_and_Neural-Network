```python
def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0.5, 'right': 0.57, 'left': 0.43}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 1.01*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')
```

```python
n_groups = 3
train_acc = np.round(acc[0:10:3],2)
valid_acc = np.round(acc[1:10:3],2)
test_acc = np.round(acc[2:10:3],2)

fig, ax = plt.subplots(figsize=(8, 6))

index = np.arange(n_groups)
bar_width = 0.2

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = ax.bar(index - bar_width, train_acc, bar_width,
                alpha=opacity, color='b', error_kw=error_config,
                label='Training data')

rects2 = ax.bar(index, valid_acc, bar_width,
                alpha=opacity, color='r', error_kw=error_config,
                label='Validation data')

rects3 = ax.bar(index + bar_width, test_acc, bar_width,
                alpha=opacity, color='g', error_kw=error_config,
                label='Testing data')

ax.set_ylabel('Accuracy(%)')
ax.set_title('SVM kernel comparison')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(('linear kernel', 'RBF(gamma=1)', 'RBF(gamma=default)'))

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
autolabel(rects1, "left")
autolabel(rects2)
autolabel(rects3, "right")


```
